```python
import pandas as pd
```

```python
HSC = pd.read_csv('../data/cell_type_specific_data/ProcessedData/HSC/regulation.csv')
MEP = pd.read_csv('../data/cell_type_specific_data/ProcessedData/MEP/regulation.csv')
ERY = pd.read_csv('../data/cell_type_specific_data/ProcessedData/ERY/regulation.csv')
```

```python
HSC = HSC[HSC['TF'].isin(['MECOM', 'KLF1'])]
MEP = MEP[MEP['TF'].isin(['MECOM', 'KLF1'])]
ERY = ERY[ERY['TF'].isin(['MECOM', 'KLF1'])]
```

```python
# reset index
HSC = HSC.reset_index(drop=True)
MEP = MEP.reset_index(drop=True)
ERY = ERY.reset_index(drop=True)
```

```python
# Concate the three dataframes
HSC['cell_type'] = 'HSC'
MEP['cell_type'] = 'MEP'
ERY['cell_type'] = 'ERY'
regulation = pd.concat([HSC, MEP, ERY], axis=0)
```

```python
regulation.reset_index(drop=True, inplace=True)
```

```python
regulation.to_csv('./_figure/cell_type_network/cell_type_network.csv', index=False)
```
