```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the csv file
df_tissue_specific_deepmix = pd.read_csv('./_doc/result_tissue_specific_deepmix.csv', index_col=0, header=0)
df_tissue_specific_cnnc = pd.read_csv('./_doc/result_tissue_specific_cnnc.csv', index_col=0, header=0)
df_tissue_specific_dgrns = pd.read_csv('./_doc/result_tissue_specific_dgrns.csv', index_col=0, header=0)
df_tissue_specific_stgrns = pd.read_csv('./_doc/result_tissue_specific_stgrns.csv', index_col=0, header=0)
df_tissue_specific_genelink = pd.read_csv('./_doc/result_tissue_specific_genelink.csv', index_col=0, header=0)
df_tissue_specific_deeptfni = pd.read_csv('./_doc/result_tissue_specific_deeptfni.csv', index_col=0, header=0)
df_tissue_specific_deepsem = pd.read_csv('./_doc/result_tissue_specific_deepsem.csv', index_col=0, header=0)


# group by the same dataset
df_tissue_specific_deepmix.insert(0, 'dataset', df_tissue_specific_deepmix.index)
df_tissue_specific_deepmix = df_tissue_specific_deepmix.drop_duplicates(subset=['dataset', 'seed', 'cell', 'size'], keep='first')
df_tissue_specific_deepmix = df_tissue_specific_deepmix[(df_tissue_specific_deepmix['seed'] == 8) & (df_tissue_specific_deepmix['cell'] == 5000)]
df_tissue_specific_deepmix = df_tissue_specific_deepmix[['size', 'auroc', 'aupr', 'aupr_norm', 'ep', 'epr']]
df_tissue_specific_deepmix = df_tissue_specific_deepmix.sort_values(by='size', ascending=True)
df_tissue_specific_cnnc.insert(0, 'dataset', df_tissue_specific_cnnc.index)
df_tissue_specific_cnnc = df_tissue_specific_cnnc.drop_duplicates(subset=['dataset', 'seed', 'cell', 'size'], keep='first')
df_tissue_specific_cnnc = df_tissue_specific_cnnc[(df_tissue_specific_cnnc['seed'] == 8) & (df_tissue_specific_cnnc['cell'] == 5000)]
df_tissue_specific_cnnc = df_tissue_specific_cnnc[['size', 'auroc', 'aupr', 'aupr_norm', 'ep', 'epr']]
df_tissue_specific_cnnc = df_tissue_specific_cnnc.sort_values(by='size', ascending=True)
df_tissue_specific_dgrns.insert(0, 'dataset', df_tissue_specific_dgrns.index)
df_tissue_specific_dgrns = df_tissue_specific_dgrns.drop_duplicates(subset=['dataset', 'seed', 'cell', 'size'], keep='first')
df_tissue_specific_dgrns = df_tissue_specific_dgrns[(df_tissue_specific_dgrns['seed'] == 8) & (df_tissue_specific_dgrns['cell'] == 5000)]
df_tissue_specific_dgrns = df_tissue_specific_dgrns[['size', 'auroc', 'aupr', 'aupr_norm', 'ep', 'epr']]
df_tissue_specific_dgrns = df_tissue_specific_dgrns.sort_values(by='size', ascending=True)
df_tissue_specific_stgrns.insert(0, 'dataset', df_tissue_specific_stgrns.index)
df_tissue_specific_stgrns = df_tissue_specific_stgrns.drop_duplicates(subset=['dataset', 'seed', 'cell', 'size'], keep='first')
df_tissue_specific_stgrns = df_tissue_specific_stgrns[(df_tissue_specific_stgrns['seed'] == 8) & (df_tissue_specific_stgrns['cell'] == 5000)]
df_tissue_specific_stgrns = df_tissue_specific_stgrns[['size', 'auroc', 'aupr', 'aupr_norm', 'ep', 'epr']]
df_tissue_specific_stgrns = df_tissue_specific_stgrns.sort_values(by='size', ascending=True)
df_tissue_specific_genelink.insert(0, 'dataset', df_tissue_specific_genelink.index)
df_tissue_specific_genelink = df_tissue_specific_genelink.drop_duplicates(subset=['dataset', 'seed', 'cell', 'size'], keep='first')
df_tissue_specific_genelink = df_tissue_specific_genelink[(df_tissue_specific_genelink['seed'] == 8) & (df_tissue_specific_genelink['cell'] == 5000)]
df_tissue_specific_genelink = df_tissue_specific_genelink[['size', 'auroc', 'aupr', 'aupr_norm', 'ep', 'epr']]
df_tissue_specific_genelink = df_tissue_specific_genelink.sort_values(by='size', ascending=True)
df_tissue_specific_deeptfni.insert(0, 'dataset', df_tissue_specific_deeptfni.index)
df_tissue_specific_deeptfni = df_tissue_specific_deeptfni.drop_duplicates(subset=['dataset', 'seed', 'cell', 'size'], keep='first')
df_tissue_specific_deeptfni = df_tissue_specific_deeptfni[(df_tissue_specific_deeptfni['seed'] == 8) & (df_tissue_specific_deeptfni['cell'] == 5000)]
df_tissue_specific_deeptfni = df_tissue_specific_deeptfni[['size', 'auroc', 'aupr', 'aupr_norm', 'ep', 'epr']]
df_tissue_specific_deeptfni = df_tissue_specific_deeptfni.sort_values(by='size', ascending=True)
df_tissue_specific_deepsem.insert(0, 'dataset', df_tissue_specific_deepsem.index)
df_tissue_specific_deepsem = df_tissue_specific_deepsem.drop_duplicates(subset=['dataset', 'seed', 'cell', 'size'], keep='first')
df_tissue_specific_deepsem = df_tissue_specific_deepsem[(df_tissue_specific_deepsem['seed'] == 8) & (df_tissue_specific_deepsem['cell'] == 5000)]
df_tissue_specific_deepsem = df_tissue_specific_deepsem[['size', 'auroc', 'aupr', 'aupr_norm', 'ep', 'epr']]
df_tissue_specific_deepsem = df_tissue_specific_deepsem.sort_values(by='size', ascending=True)

print(df_tissue_specific_deepmix.shape)
print(df_tissue_specific_cnnc.shape)
print(df_tissue_specific_dgrns.shape)
print(df_tissue_specific_stgrns.shape)
print(df_tissue_specific_genelink.shape)
print(df_tissue_specific_deeptfni.shape)
print(df_tissue_specific_deepsem.shape)

# df_tissue_specific_cnnc
```

```python
# create the figure folder
os.makedirs('./_figure/model_comparison_tissue_specific', exist_ok=True)
```

```python
def plot_model_comparison_tissue_specific_with_size(metric='auroc'):
    fig, axs = plt.subplots(1, 5, figsize=(18, 5))

    axs[0].plot([1,2,3,4,5], df_tissue_specific_deepmix.loc[['tissue_human_blood']][metric], marker='v', label='SEEGRN (ours)')
    axs[0].plot([1,2,3,4,5], df_tissue_specific_stgrns.loc[['tissue_human_blood']][metric], marker='s', label='STGRNS')
    axs[0].plot([1,2,3,4,5], df_tissue_specific_genelink.loc[['tissue_human_blood']][metric], marker='d', label='GENELink')
    axs[0].plot([1,2,3,4,5], df_tissue_specific_deeptfni.loc[['tissue_human_blood']][metric], marker='^', label='DeepTFNI')
    axs[0].plot([1,2,3,4,5], df_tissue_specific_cnnc.loc[['tissue_human_blood']][metric], marker='*', label='CNCC')
    axs[0].plot([1,2,3,4,5], df_tissue_specific_dgrns.loc[['tissue_human_blood']][metric], marker='o', label='DGRNS')
    axs[0].plot([1,2,3,4,5], df_tissue_specific_deepsem.loc[['tissue_human_blood']][metric], marker='x', label='DeepSEM')
    axs[0].set_title('Human Blood', fontsize=15)
    axs[0].set_xlabel('Data Size', fontsize=15)
    axs[0].set_ylabel(metric.upper(), fontsize=15)
    axs[0].set_xticks([1,2,3,4,5])
    axs[0].set_xticklabels(df_tissue_specific_deepsem.loc[['tissue_human_blood']]['size'])

    axs[1].plot([1,2,3,4,5], df_tissue_specific_deepmix.loc[['tissue_human_bone_marrow']][metric], marker='v', label='SEEGRN (ours)')
    axs[1].plot([1,2,3,4,5], df_tissue_specific_stgrns.loc[['tissue_human_bone_marrow']][metric], marker='s', label='STGRNS')
    axs[1].plot([1,2,3,4,5], df_tissue_specific_genelink.loc[['tissue_human_bone_marrow']][metric], marker='d', label='GENELink')
    axs[1].plot([1,2,3,4,5], df_tissue_specific_deeptfni.loc[['tissue_human_bone_marrow']][metric], marker='^', label='DeepTFNI')
    axs[1].plot([1,2,3,4,5], df_tissue_specific_cnnc.loc[['tissue_human_bone_marrow']][metric], marker='*', label='CNCC')
    axs[1].plot([1,2,3,4,5], df_tissue_specific_dgrns.loc[['tissue_human_bone_marrow']][metric], marker='o', label='DGRNS')
    axs[1].plot([1,2,3,4,5], df_tissue_specific_deepsem.loc[['tissue_human_bone_marrow']][metric], marker='x', label='DeepSEM')
    axs[1].set_title('Human Bone Marrow', fontsize=15)
    axs[1].set_xlabel('Data Size', fontsize=15)
    axs[1].set_xticks([1,2,3,4,5])
    axs[1].set_xticklabels(df_tissue_specific_deepsem.loc[['tissue_human_bone_marrow']]['size'])

    axs[2].plot([1,2,3,4,5], df_tissue_specific_deepmix.loc[['tissue_human_cerebral_cortex']][metric], marker='v', label='SEEGRN (ours)')
    axs[2].plot([1,2,3,4,5], df_tissue_specific_stgrns.loc[['tissue_human_cerebral_cortex']][metric], marker='s', label='STGRNS')
    axs[2].plot([1,2,3,4,5], df_tissue_specific_genelink.loc[['tissue_human_cerebral_cortex']][metric], marker='d', label='GENELink')
    axs[2].plot([1,2,3,4,5], df_tissue_specific_deeptfni.loc[['tissue_human_cerebral_cortex']][metric], marker='^', label='DeepTFNI')
    axs[2].plot([1,2,3,4,5], df_tissue_specific_cnnc.loc[['tissue_human_cerebral_cortex']][metric], marker='*', label='CNCC')
    axs[2].plot([1,2,3,4,5], df_tissue_specific_dgrns.loc[['tissue_human_cerebral_cortex']][metric], marker='o', label='DGRNS')
    axs[2].plot([1,2,3,4,5], df_tissue_specific_deepsem.loc[['tissue_human_cerebral_cortex']][metric], marker='x', label='DeepSEM')
    axs[2].set_title('Human Cerebral Cortex', fontsize=15)
    axs[2].set_xlabel('Data Size', fontsize=15)
    axs[2].set_xticks([1,2,3,4,5])
    axs[2].set_xticklabels(df_tissue_specific_deepsem.loc[['tissue_human_cerebral_cortex']]['size'])

    axs[3].plot([1,2,3,4,5], df_tissue_specific_deepmix.loc[['tissue_mouse_brain_cortex']][metric], marker='v', label='SEEGRN (ours)')
    axs[3].plot([1,2,3,4,5], df_tissue_specific_stgrns.loc[['tissue_mouse_brain_cortex']][metric], marker='s', label='STGRNS')
    axs[3].plot([1,2,3,4,5], df_tissue_specific_genelink.loc[['tissue_mouse_brain_cortex']][metric], marker='d', label='GENELink')
    axs[3].plot([1,2,3,4,5], df_tissue_specific_deeptfni.loc[['tissue_mouse_brain_cortex']][metric], marker='^', label='DeepTFNI')
    axs[3].plot([1,2,3,4,5], df_tissue_specific_cnnc.loc[['tissue_mouse_brain_cortex']][metric], marker='*', label='CNCC')
    axs[3].plot([1,2,3,4,5], df_tissue_specific_dgrns.loc[['tissue_mouse_brain_cortex']][metric], marker='o', label='DGRNS')
    axs[3].plot([1,2,3,4,5], df_tissue_specific_deepsem.loc[['tissue_mouse_brain_cortex']][metric], marker='x', label='DeepSEM')
    axs[3].set_title('Mouse Brain Cortex', fontsize=15)
    axs[3].set_xlabel('Data Size', fontsize=15)
    axs[3].set_xticks([1,2,3,4,5])
    axs[3].set_xticklabels(df_tissue_specific_deepsem.loc[['tissue_mouse_brain_cortex']]['size'])

    axs[4].plot([1,2,3,4,5], df_tissue_specific_deepmix.loc[['tissue_mouse_skin']][metric], marker='v', label='SEEGRN (ours)')
    axs[4].plot([1,2,3,4,5], df_tissue_specific_stgrns.loc[['tissue_mouse_skin']][metric], marker='s', label='STGRNS')
    axs[4].plot([1,2,3,4,5], df_tissue_specific_genelink.loc[['tissue_mouse_skin']][metric], marker='d', label='GENELink')
    axs[4].plot([1,2,3,4,5], df_tissue_specific_deeptfni.loc[['tissue_mouse_skin']][metric], marker='^', label='DeepTFNI')
    axs[4].plot([1,2,3,4,5], df_tissue_specific_cnnc.loc[['tissue_mouse_skin']][metric], marker='*', label='CNCC')
    axs[4].plot([1,2,3,4,5], df_tissue_specific_dgrns.loc[['tissue_mouse_skin']][metric], marker='o', label='DGRNS')
    axs[4].plot([1,2,3,4,5], df_tissue_specific_deepsem.loc[['tissue_mouse_skin']][metric], marker='x', label='DeepSEM')
    axs[4].set_title('Mouse Skin', fontsize=15)
    axs[4].set_xlabel('Data Size', fontsize=15)
    axs[4].set_xticks([1,2,3,4,5])
    axs[4].set_xticklabels(df_tissue_specific_deepsem.loc[['tissue_mouse_skin']]['size'])
    axs[4].legend(loc='lower right', bbox_to_anchor=(1.70, 0.0), fontsize=10, borderaxespad=0)

    plt.tight_layout()
    plt.savefig('./_figure/model_comparison_tissue_specific/lineplot_tissue_specific_{0}_size.jpg'.format(metric), dpi=100)
    plt.savefig('./_figure/model_comparison_tissue_specific/lineplot_tissue_specific_{0}_size.pdf'.format(metric), dpi=100)
    plt.close()
```

```python
plot_model_comparison_tissue_specific_with_size('auroc')
plot_model_comparison_tissue_specific_with_size('aupr')
plot_model_comparison_tissue_specific_with_size('aupr_norm')
plot_model_comparison_tissue_specific_with_size('ep')
plot_model_comparison_tissue_specific_with_size('epr')
```
