# Analysis Sequence Pretrain Model

```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# plt.style.use('seaborn-whitegrid')
# plt.style.use('seaborn-v0_8-whitegrid')
plt.style.use('seaborn-colorblind')

# Plot two subplots left and right
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 5))

# Plot the first subplot
# Plot a grouped bar chart
# Set the bar width
barWidth = 0.20
# Set the position of the bar on the x-axis
r1 = np.arange(4)
r2 = [x + barWidth + 0.05 for x in r1]
# Make the plot
homo_result_roc = [0.5, 0.948, 0.641, 0.940]
mus_result_roc = [0.5, 0.686, 0.962, 0.949]
reacts = ax1.bar(r1, homo_result_roc, width=barWidth, label='Homo Sapiens')
ax1.bar_label(reacts, padding=5, rotation=-45)
reacts = ax1.bar(r2, mus_result_roc, width=barWidth, label='Mus Musculus')
ax1.bar_label(reacts, padding=5, rotation=-45)
# Add xticks on the middle of the group bars
ax1.set_xlabel('Model', fontweight='bold')
ax1.set_ylabel('Value', fontweight='bold')
ax1.set_xticks([r + barWidth/2 for r in range(4)])
ax1.set_xticklabels(['Baseline', 'Homo Sapiens Model', 'Mus Musculus Model', 'Cross Species Model'])
ax1.set_ylim(0, 1.1)
ax1.set_title('AUROC', pad=20, fontdict={'fontsize': 15, 'fontweight': 'bold'})
ax1.legend(loc='upper left')
# disable the grid
ax1.grid(False)

# Plot the second subplot
# Plot a grouped bar chart
# Set the bar width
barWidth = 0.20
# Set the position of the bar on the x-axis
r1 = np.arange(4)
r2 = [x + barWidth + 0.05 for x in r1]
# Make the plot
homo_result_prc = [0.218, 0.851, 0.377, 0.828]
mus_result_prc = [0.175, 0.362, 0.853, 0.804]
reacts = ax2.bar(r1, homo_result_prc, width=barWidth, label='Homo Sapiens')
ax2.bar_label(reacts, padding=5, rotation=-45)
reacts = ax2.bar(r2, mus_result_prc, width=barWidth, label='Mus Musculus')
ax2.bar_label(reacts, padding=5, rotation=-45)
# Add xticks on the middle of the group bars
ax2.set_xlabel('Model', fontweight='bold')
ax2.set_ylabel('Value', fontweight='bold')
ax2.set_xticks([r + barWidth/2 for r in range(4)])
ax2.set_xticklabels(['Baseline', 'Homo Sapiens Model', 'Mus Musculus Model', 'Cross Species Model'])

# set the y-axis ticks

ax2.set_ylim(0, 1.1)
ax2.set_title('AUPR', pad=20, fontdict={'fontsize': 15, 'fontweight': 'bold'})
ax2.legend(loc='upper left')
# disable the grid
ax2.grid(False)

# Create legend & Show graphic

# plt.show()
os.makedirs('./_figure/model_pretrain_result', exist_ok=True)
plt.savefig('./_figure/model_pretrain_result/sequence_pretrain_result.pdf', dpi=600)
plt.savefig('./_figure/model_pretrain_result/sequence_pretrain_result.jpg', dpi=100)
```
