```python
# Read the csv file
df_cell_type_specific_deepmix = pd.read_csv('./_doc/result_cell_type_specific_deepmix.csv', index_col=0, header=0)
df_cell_type_specific_cnnc = pd.read_csv('./_doc/result_cell_type_specific_cnnc.csv', index_col=0, header=0)
df_cell_type_specific_dgrns = pd.read_csv('./_doc/result_cell_type_specific_dgrns.csv', index_col=0, header=0)
df_cell_type_specific_stgrns = pd.read_csv('./_doc/result_cell_type_specific_stgrns.csv', index_col=0, header=0)
df_cell_type_specific_genelink = pd.read_csv('./_doc/result_cell_type_specific_genelink.csv', index_col=0, header=0)
df_cell_type_specific_deeptfni = pd.read_csv('./_doc/result_cell_type_specific_deeptfni.csv', index_col=0, header=0)
df_cell_type_specific_deepsem = pd.read_csv('./_doc/result_cell_type_specific_deepsem.csv', index_col=0, header=0)


# group by the same dataset
df_cell_type_specific_deepmix.insert(0, 'dataset', df_cell_type_specific_deepmix.index)
df_cell_type_specific_deepmix = df_cell_type_specific_deepmix.drop_duplicates(subset=['dataset', 'seed', 'cell', 'size'], keep='first')
df_cell_type_specific_deepmix = df_cell_type_specific_deepmix[(df_cell_type_specific_deepmix['cell'] == 5000) & (df_cell_type_specific_deepmix['size'] == 1.0)]
df_cell_type_specific_deepmix = df_cell_type_specific_deepmix[['auroc', 'aupr', 'aupr_norm', 'ep', 'epr']]
df_cell_type_specific_cnnc.insert(0, 'dataset', df_cell_type_specific_cnnc.index)
df_cell_type_specific_cnnc = df_cell_type_specific_cnnc.drop_duplicates(subset=['dataset', 'seed', 'cell', 'size'], keep='first')
df_cell_type_specific_cnnc = df_cell_type_specific_cnnc[(df_cell_type_specific_cnnc['cell'] == 5000) & (df_cell_type_specific_cnnc['size'] == 1.0)]
df_cell_type_specific_cnnc = df_cell_type_specific_cnnc[['auroc', 'aupr', 'aupr_norm', 'ep', 'epr']]
df_cell_type_specific_dgrns.insert(0, 'dataset', df_cell_type_specific_dgrns.index)
df_cell_type_specific_dgrns = df_cell_type_specific_dgrns.drop_duplicates(subset=['dataset', 'seed', 'cell', 'size'], keep='first')
df_cell_type_specific_dgrns = df_cell_type_specific_dgrns[(df_cell_type_specific_dgrns['cell'] == 5000) & (df_cell_type_specific_dgrns['size'] == 1.0)]
df_cell_type_specific_dgrns = df_cell_type_specific_dgrns[['auroc', 'aupr', 'aupr_norm', 'ep', 'epr']]
df_cell_type_specific_stgrns.insert(0, 'dataset', df_cell_type_specific_stgrns.index)
df_cell_type_specific_stgrns = df_cell_type_specific_stgrns.drop_duplicates(subset=['dataset', 'seed', 'cell', 'size'], keep='first')
df_cell_type_specific_stgrns = df_cell_type_specific_stgrns[(df_cell_type_specific_stgrns['cell'] == 5000) & (df_cell_type_specific_stgrns['size'] == 1.0)]
df_cell_type_specific_stgrns = df_cell_type_specific_stgrns[['auroc', 'aupr', 'aupr_norm', 'ep', 'epr']]
df_cell_type_specific_genelink.insert(0, 'dataset', df_cell_type_specific_genelink.index)
df_cell_type_specific_genelink = df_cell_type_specific_genelink.drop_duplicates(subset=['dataset', 'seed', 'cell', 'size'], keep='first')
df_cell_type_specific_genelink = df_cell_type_specific_genelink[(df_cell_type_specific_genelink['cell'] == 5000) & (df_cell_type_specific_genelink['size'] == 1.0)]
df_cell_type_specific_genelink = df_cell_type_specific_genelink[['auroc', 'aupr', 'aupr_norm', 'ep', 'epr']]
df_cell_type_specific_deeptfni.insert(0, 'dataset', df_cell_type_specific_deeptfni.index)
df_cell_type_specific_deeptfni = df_cell_type_specific_deeptfni.drop_duplicates(subset=['dataset', 'seed', 'cell', 'size'], keep='first')
df_cell_type_specific_deeptfni = df_cell_type_specific_deeptfni[(df_cell_type_specific_deeptfni['cell'] == 5000) & (df_cell_type_specific_deeptfni['size'] == 1.0)]
df_cell_type_specific_deeptfni = df_cell_type_specific_deeptfni[['auroc', 'aupr', 'aupr_norm', 'ep', 'epr']]
df_cell_type_specific_deepsem.insert(0, 'dataset', df_cell_type_specific_deepsem.index)
df_cell_type_specific_deepsem = df_cell_type_specific_deepsem.drop_duplicates(subset=['dataset', 'seed', 'cell', 'size'], keep='first')
df_cell_type_specific_deepsem = df_cell_type_specific_deepsem[(df_cell_type_specific_deepsem['cell'] == 5000) & (df_cell_type_specific_deepsem['size'] == 1.0)]
df_cell_type_specific_deepsem = df_cell_type_specific_deepsem[['auroc', 'aupr', 'aupr_norm', 'ep', 'epr']]

print(df_cell_type_specific_deepmix.shape)
print(df_cell_type_specific_cnnc.shape)
print(df_cell_type_specific_dgrns.shape)
print(df_cell_type_specific_stgrns.shape)
print(df_cell_type_specific_genelink.shape)
print(df_cell_type_specific_deeptfni.shape)
print(df_cell_type_specific_deepsem.shape)
```

```python
# beautiful colors that are suitable for scholars
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']


# create the figure folder
os.makedirs('./_figure/model_comparison_cell_type_specific', exist_ok=True)
```

```python
def plot_model_comparison_cell_type_specific_with_seed_main(metric='auroc', df_cell_type_specific=df_cell_type_specific_auroc):
    fig = plt.figure(figsize=(12, 6))
    labels = df_cell_type_specific.columns
    labels = [label.replace('DeepMix', 'seeGRN (ours)') for label in labels]

    boxplot_1 = plt.boxplot(df_cell_type_specific.loc[['cell_type_ery']].values, patch_artist=True, labels=labels, positions=[1, 2, 3, 4, 5, 6, 7], widths=0.6, showfliers=False)
    plt.setp(boxplot_1['medians'], color='black')
    plt.setp(boxplot_1['whiskers'], color='black')
    for patch, color in zip(boxplot_1['boxes'], colors):
        # patch.set_alpha(0.5) # set the transparency of the box
        patch.set_facecolor(color)

    boxplot_2 = plt.boxplot(df_cell_type_specific.loc[['cell_type_hsc']].values, patch_artist=True, labels=labels, positions=[9, 10, 11, 12, 13, 14, 15], widths=0.6, showfliers=False)
    plt.setp(boxplot_2['medians'], color='black')
    plt.setp(boxplot_2['whiskers'], color='black')
    for patch, color in zip(boxplot_2['boxes'], colors):
        patch.set_facecolor(color)

    boxplot_3 = plt.boxplot(df_cell_type_specific.loc[['cell_type_mep']].values, patch_artist=True, labels=labels, positions=[17, 18, 19, 20, 21, 22, 23], widths=0.6, showfliers=False)
    plt.setp(boxplot_3['medians'], color='black')
    plt.setp(boxplot_3['whiskers'], color='black')
    for patch, color in zip(boxplot_3['boxes'], colors):
        patch.set_facecolor(color)

    x_position=[4, 12, 20]
    x_position_fmt=['cell_type_ery', 'cell_type_hsc', 'cell_type_mep']
    plt.xticks(x_position, x_position_fmt, fontsize=15)
    plt.xlim(0, 24)
    plt.yticks(fontsize=15)
    ymin, ymax = plt.ylim()
    plt.ylim(ymin, ymax)
    plt.ylabel(metric.upper(), fontsize=20)
    # plt.legend(boxplot_1["boxes"], df_cell_type_specific.columns, loc='lower right', fontsize=12) 
    plt.legend(boxplot_1["boxes"], labels, loc='lower right', bbox_to_anchor=(1.20, 0.0), fontsize=10, borderaxespad=0) # borderaxespad 

    # plt vertical dotted line between four boxplots
    plt.plot([8, 8], [ymin, ymax], 'k--', lw=1)
    plt.plot([16, 16], [ymin, ymax], 'k--', lw=1)
    plt.plot([24, 24], [ymin, ymax], 'k--', lw=1)

    plt.tight_layout()
    plt.savefig('./_figure/model_comparison_cell_type_specific/boxplot_cell_type_specific_{}_seed_main.jpg'.format(metric), dpi=100, bbox_inches='tight')
    plt.savefig('./_figure/model_comparison_cell_type_specific/boxplot_cell_type_specific_{}_seed_main.pdf'.format(metric), dpi=100, bbox_inches='tight')
    plt.close()
```

```python
plot_model_comparison_cell_type_specific_with_seed_main('auroc', df_cell_type_specific=df_cell_type_specific_auroc)
plot_model_comparison_cell_type_specific_with_seed_main('aupr', df_cell_type_specific=df_cell_type_specific_aupr)
plot_model_comparison_cell_type_specific_with_seed_main('aupr_norm', df_cell_type_specific=df_cell_type_specific_aupr_norm)
plot_model_comparison_cell_type_specific_with_seed_main('ep', df_cell_type_specific=df_cell_type_specific_ep)
plot_model_comparison_cell_type_specific_with_seed_main('epr', df_cell_type_specific=df_cell_type_specific_epr)
```
