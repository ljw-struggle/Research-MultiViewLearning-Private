```python
import os
import re
import csv
import numpy as np
import pandas as pd
```

```python
def collect_ablation_result(tissue_result_dir, tissue):
    final_result_file = os.path.join(tissue_result_dir, 'ablation_result_{}.csv'.format(tissue))

    # Get the result dir list in the dir
    result_list = []
    tissue_result_dir = os.path.join(tissue_result_dir, tissue)
    result_dir_list = os.listdir(tissue_result_dir)

    for result_dir in result_dir_list:
        result_name = result_dir
        result_dir = os.path.join(tissue_result_dir, result_dir)
        result_file = os.path.join(result_dir, 'ORIGIN/log/info.log')
        if os.path.exists(result_file):
            # Get the last line of the result file
            with open(result_file, 'r') as f:
                lines = f.readlines()
                last_line = lines[-1]
                try:
                    result = eval(last_line.split(' - INFO - ')[-1])
                    result['result_name'] = result_name
                    result_list.append(result)
                except:
                    print('Error: ' + result_file)

    # Write the result list to the final result csv file
    with open(final_result_file, 'w') as f:
        csv_writer = csv.writer(f)
        csv_writer.writerow(['result_name', 'accuracy', 'precision', 'recall', 'f1', 'auroc', 'aupr', 'aupr_norm', 'ep', 'epr', 'macs', 'params', 'memory'])
        for result in result_list:
            csv_writer.writerow([result['result_name'], result['balanced_accuracy'], result['precision'], result['recall'], result['f1'], result['auroc'], result['aupr'], result['ep'], result['epr'], result['macs'], result['params']])
    
    return result_list
```

```python
tissue_result_dir = './DeepMixModel/tissue_specific_data/'
result_tissue_human_blood = collect_ablation_result(tissue_result_dir, 'Human-Blood')
result_tissue_human_bone_marrow = collect_ablation_result(tissue_result_dir, 'Human-Bone-Marrow')
result_tissue_human_cerebral_cortex = collect_ablation_result(tissue_result_dir, 'Human-Cerebral-Cortex')
result_tissue_mouse_brain_cortex = collect_ablation_result(tissue_result_dir, 'Mouse-Brain-Cortex')
result_tissue_mouse_skin = collect_ablation_result(tissue_result_dir, 'Mouse-Skin')
```
