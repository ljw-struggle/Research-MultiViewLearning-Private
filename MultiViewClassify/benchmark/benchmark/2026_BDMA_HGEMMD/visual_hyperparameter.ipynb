```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import FormatStrFormatter, MultipleLocator

temp_num = 1
for data in ['BRCA', 'ROSMAP']:
    df = pd.read_csv(f'./figure/data_MOGONET/hp_l1_l2_{data}.csv')
    lambda_1_values = df['l1'].unique()
    lambda_2_values = df['l2'].unique()
    lambda_1_grid, lambda_2_grid = np.meshgrid(lambda_1_values, lambda_2_values)
    lambda_1_flat = lambda_1_grid.flatten()
    lambda_2_flat = lambda_2_grid.flatten()
    acc_dict = dict(zip(zip(df['l1'], df['l2']), df['ACC']))
    acc_values = np.array([acc_dict[(l1, l2)] for l1, l2 in zip(lambda_1_flat, lambda_2_flat)])
    # acc_values = acc_values - 0.7
    acc_values = (acc_values - 0.7) * 100
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    colormap = plt.get_cmap('viridis')  # You can change the colormap
    # norm = plt.Normalize(vmin=0, vmax=0.2)  # Normalize values to map to colormap
    norm = plt.Normalize(vmin=0, vmax=20)  # Normalize values to map to colormap
    colors = colormap(norm(acc_values))  # Get colors based on the normalized 'acc' values
    max_idx = np.argmax(acc_values)
    colors[max_idx] = [1.0, 0.0, 0.0, 1.0]  # RGBA red 
    max_l1 = lambda_1_flat[max_idx]
    max_l2 = lambda_2_flat[max_idx]
    max_acc = acc_values[max_idx]
    bars = ax.bar3d(lambda_1_flat, lambda_2_flat, np.zeros_like(acc_values), 0.06, 0.06, acc_values, color=colors, edgecolor='k', alpha=1, zsort='average')
    # ax.text(max_l1, max_l2, max_acc + 0.01, f'{max_acc + 0.7:.3f}', color='red', fontsize=10, weight='bold', ha='center', zorder=10)
    ax.text(max_l1, max_l2, max_acc + 1, f'{max_acc + 70:.1f}', color='red', fontsize=10, weight='bold', ha='center', zorder=10)
    # ax.scatter(max_l1 + 0.03, max_l2 + 0.03, max_acc + 0.02, color='red', s=0, label='Max ACC', zorder=1000, depthshade=False)
    ax.invert_yaxis() 
    ax.set_xticks(lambda_1_values)
    # ax.set_xticklabels(lambda_1_values, fontsize=9, fontdict={'weight': 'normal', 'size': 10, 'style': 'italic'}, ha='center')
    ax.set_xticklabels(lambda_1_values, fontsize=9, fontdict={'weight': 'normal', 'size': 10}, ha='center')
    ax.set_yticks(lambda_2_values)
    # ax.set_yticklabels(lambda_2_values, fontsize=9, fontdict={'weight': 'normal', 'size': 10, 'style': 'italic'}, ha='center')
    ax.set_yticklabels(lambda_2_values, fontsize=9, fontdict={'weight': 'normal', 'size': 10}, ha='center')
    # ax.set_zlim(0, 0.21)
    ax.set_zlim(0, 21)
    # ax.set_zticks(np.arange(0, 0.21, 0.05))
    ax.set_zticks(np.arange(0, 21, 5))
    # ax.set_zticklabels(np.round(np.arange(0.7, 0.91, 0.05), 2), fontsize=9, fontdict={'weight': 'normal', 'size': 10, 'style': 'italic'}, ha='center')
    # z_ticks = np.arange(0.7, 0.91, 0.05)
    z_ticks = np.arange(70, 91, 5)
    # z_ticklabels = [f"{x:.2f}" for x in z_ticks]
    z_ticklabels = [f"{x:.0f}" for x in z_ticks]
    # ax.set_zticklabels(z_ticklabels, fontsize=9, fontdict={'weight': 'normal', 'size': 10, 'style': 'italic'}, ha='center')
    ax.set_zticklabels(z_ticklabels, fontsize=9, fontdict={'weight': 'normal', 'size': 10}, ha='center')
    ax.set_xlabel(r'$\lambda_1$')
    ax.set_ylabel(r'$\lambda_2$')
    # ax.set_zlabel('Accuracy', fontsize=10, fontdict={'weight': 'normal', 'size': 10, 'style': 'italic'})
    ax.set_zlabel('Accuracy', fontsize=10, fontdict={'weight': 'normal', 'size': 10})
    ax.set_title('')
    # ax.legend()
    # mappable = plt.cm.ScalarMappable(cmap=colormap, norm=plt.Normalize(vmin=0.7, vmax=0.9))
    mappable = plt.cm.ScalarMappable(cmap=colormap, norm=plt.Normalize(vmin=70, vmax=90))
    mappable.set_array(acc_values)
    cbar = fig.colorbar(mappable, ax=ax, label='', shrink=0.5, pad=0.15)
    cbar.ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
    cbar.ax.yaxis.set_major_locator(MultipleLocator(10))   
    cbar.ax.set_ylabel('Accuracy', fontsize=10, fontdict={'weight': 'normal', 'size': 10})
    # plt.tight_layout()
    plt.savefig(f'./figure/data_MOGONET/hyper_{temp_num}.png', dpi=400, bbox_inches='tight')
    temp_num += 1
    plt.show()
```
