```python
import os
import pandas as pd
import matplotlib.pyplot as plt

temp_num = 1
colors = {'HGEMMD': '#1f77b4', 'GEMMD_GCN': '#ff7f0e', 'MMD': '#2ca02c'}
for data in ['BRCA', 'ROSMAP']:
    # metrics = ['ACC', 'F1_weighted'] if data in ['BRCA'] else ['ACC', 'AUC']
    metrics = ['ACC', 'F1_weighted'] if data in ['BRCA'] else ['ACC', 'AUC']
    for metric in metrics:
        plot_data = pd.read_csv('./figure/data_MOGONET/noisy_condition.csv')
        plot_data = plot_data[(plot_data['data'] == data) & (plot_data['metric'] == metric)]
        plot_data = plot_data[['method', 'sigma', 'mean', 'std']]
        plot_data['method'] = pd.Categorical(plot_data['method'], categories=['HGEMMD', 'GEMMD_GCN', 'MMD'], ordered=True)
        plot_data = plot_data.sort_values(by=['method', 'sigma']).reset_index(drop=True)
        plt.figure(figsize=(5, 4))
        for method, g in plot_data.groupby('method'):
        # for method, g in plot_data.groupby('method', observed=True):
            color = colors[method]
            new_method = method.replace('GEMMD_GCN', 'GEMMD (GCN)')
            plt.plot(g['sigma'], g['mean'], label=new_method, marker='o', color=color)
            plt.fill_between(g['sigma'], g['mean'] - g['std'], g['mean'] + g['std'], alpha=0.5, color=color)
        # plt.xlabel(r'Noise Level ($\sigma$)', fontsize=12, fontstyle='italic', color='black')
        plt.xlabel(r'Noise Level ($\sigma$)', fontsize=12, fontweight='bold', color='black')
        # plt.xticks(np.arange(0.0, 1.1, 0.1), [f'{i:.1f}' for i in np.arange(0.0, 1.1, 0.1)], fontsize=11, fontweight='bold', fontstyle='italic', color='black')
        plt.xticks(np.arange(0.0, 1.1, 0.1), [f'{i:.1f}' for i in np.arange(0.0, 1.1, 0.1)], fontsize=11, fontweight='bold', color='black')
        # plt.yticks(np.arange(50, 101, 5), fontsize=11, fontweight='bold', fontstyle='italic', color='black')
        plt.yticks(np.arange(50, 101, 5), fontsize=11, fontweight='bold', color='black')
        # plt.ylabel(f'{metric}', fontsize=12, fontweight='bold', fontstyle='italic', color='black')
        metric = {'ACC': 'Accuracy (%)', 'F1_weighted': 'Weighted F1 (%)', 'AUC': 'AUC (%)'}[metric]
        plt.ylabel(f'{metric}', fontsize=12, fontweight='bold', color='black')
        # plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=3, fontsize=11, frameon=False, prop={'style': 'italic', 'weight': 'bold'})
        plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=3, fontsize=11, frameon=False, prop={'weight': 'bold'})
        plt.grid(True)
        plt.tight_layout()
        plt.savefig(f'./figure/data_MOGONET/noisy_condition_{temp_num}.png', dpi=400, bbox_inches='tight'); temp_num += 1
        plt.show()
```
