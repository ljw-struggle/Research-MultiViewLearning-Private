```python
import plotly.graph_objects as go
import numpy as np

np.random.seed(42)
confusion_matrix = np.random.randint(0, 100, size=(10, 10))
confusion_matrix = confusion_matrix + np.eye(10) * 1000
```

```python
import plotly.graph_objects as go
import numpy as np

# 定义细胞类型标签
cell_types = ["T cells", "B cells", "Monocytes", "NK cells", "Dendritic cells", 
              "Neutrophils", "Eosinophils", "Basophils", "Megakaryocytes", "Erythrocytes"]

# 使用之前生成的混淆矩阵
conf_matrix = confusion_matrix

# 定义每种细胞类型的颜色
cell_colors = ['rgba(31, 119, 180, 0.8)', 'rgba(255, 127, 14, 0.8)', 'rgba(44, 160, 44, 0.8)', 
               'rgba(214, 39, 40, 0.8)', 'rgba(148, 103, 189, 0.8)', 'rgba(140, 86, 75, 0.8)', 
               'rgba(227, 119, 194, 0.8)', 'rgba(127, 127, 127, 0.8)', 'rgba(188, 189, 34, 0.8)', 
               'rgba(23, 190, 207, 0.8)']

# 创建source和target的索引
source = []
target = []
value = []
link_color = []

# 定义透明度
opacity = 0.4

# 将混淆矩阵的数据转换为桑基图所需的格式
max_value = conf_matrix.max()  # 用于归一化颜色
for i in range(len(cell_types)):  # 对每个真实细胞类型
    for j in range(len(cell_types)):  # 对每个预测细胞类型
        if conf_matrix[i, j] > 0:  # 如果流动数值大于0，则添加到桑基图
            source.append(i)  # 真实标签的索引
            target.append(len(cell_types) + j)  # 预测标签的索引（区分开真实标签）
            value.append(conf_matrix[i, j])  # 流动的数量
            # 颜色保持一致
            link_color.append(cell_colors[i])

# 创建桑基图
labels = cell_types + ["Pred_" + cell_type for cell_type in cell_types]
node_colors = cell_colors + cell_colors  # 确保预测标签颜色和真实标签一致

# 自定义桑基图为上下布局
fig = go.Figure(data=[go.Sankey(
    arrangement="snap",
    orientation="v",  # 设置为垂直方向的流动
    valueformat = ".0f",
    valuesuffix = " cells",
    node=dict(
        pad=10,
        thickness=5,
        line=dict(color="black", width=1),
        label=labels,
        color=node_colors,
        align="left",
    ),
    link=dict(
        source=source,  # 来源节点
        target=target,  # 目标节点
        value=value,    # 流动权重
        color=link_color  # 流动线的颜色
    )
)])

# 更新图表布局为上下布局
fig.update_layout(title_text="Cell Type Flow from True to Predicted Labels (Vertical Layout)", font_size=10, height=300, width=800)

# 展示图表
fig.show()
```
