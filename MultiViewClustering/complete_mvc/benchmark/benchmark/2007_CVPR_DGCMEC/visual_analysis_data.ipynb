```python
import numpy as np
from sklearn.cluster import KMeans
from _utils import MMDataset, clustering_acc, purity_score

dataset = MMDataset('./data/data_sc_multiomics/TEA/', concat_data=False)
data = [x.clone().numpy() for x in dataset.X]; label = dataset.Y.clone().numpy()
n_clusters = dataset.categories
print(f"Number of clusters: {n_clusters}")
modality_names = ['RNA', 'Protein', 'ATAC']
for i, modality_data in enumerate(data):
    kmeans = KMeans(n_clusters=n_clusters, n_init=20, random_state=0)
    preds = kmeans.fit_predict(modality_data)
    acc = clustering_acc(label, preds)
    pur = purity_score(label, preds)
    print(f"{modality_names[i]}: ACC={acc:.4f}, Purity={pur:.4f}")

concatenated_data = np.concatenate(data, axis=1)
kmeans_concat = KMeans(n_clusters=n_clusters, n_init=20, random_state=0)
preds_concat = kmeans_concat.fit_predict(concatenated_data)
acc_concat = clustering_acc(label, preds_concat)
pur_concat = purity_score(label, preds_concat)
print(f"Concatenated: ACC={acc_concat:.4f}, Purity={pur_concat:.4f}")
```
