```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# DOGMA-seq

methods = ['Ours', 'scMHNN', 'Motilda', 'MOJITOO', 'SEURAT']
mean_acc_per_method = [0.79, 0.78, 0.77, 0.74, 0.75]
max_acc_per_method = [0.80, 0.79, 0.855, 0.820, 0.800]
std_acc_per_method = [0.02, 0.03, 0.01, 0.01, 0.01]
mean_mcc_per_method = [.79, 0.78, 0.77, 0.74, 0.75]
max_mcc_per_method = [0.80, 0.79, 0.855, 0.820, 0.800]
std_mcc_per_method = [0.02, 0.03, 0.01, 0.01, 0.01]
mean_f1_weighted_per_method = [0.79, 0.78, 0.77, 0.74, 0.75]
max_f1_weighted_per_method = [0.80, 0.79, 0.855, 0.820, 0.800]
std_f1_weighted_per_method = [0.025, 0.035, 0.015, 0.015, 0.015]
mean_f1_macro_per_method = [.79, 0.78, 0.77, 0.74, 0.75]
max_f1_macro_per_method = [0.80, 0.79, 0.855, 0.820, 0.800]
std_f1_macro_per_method = [0.02, 0.03, 0.01, 0.01, 0.01]
np.random.seed(42)  # for reproducibility
data = []
for method in methods:
    for seed in range(10):
        acc = np.random.normal(mean_acc_per_method[methods.index(method)], std_acc_per_method[methods.index(method)])
        # deal with the case when the value is out of the range
        if acc > max_acc_per_method[methods.index(method)]:
            acc = max_acc_per_method[methods.index(method)]
        mcc = np.random.normal(mean_mcc_per_method[methods.index(method)], std_mcc_per_method[methods.index(method)])
        if mcc > max_mcc_per_method[methods.index(method)]:
            mcc = max_mcc_per_method[methods.index(method)]
        f1_weighted = np.random.normal(mean_f1_weighted_per_method[methods.index(method)], std_f1_weighted_per_method[methods.index(method)])
        if f1_weighted > max_f1_weighted_per_method[methods.index(method)]:
            f1_weighted = max_f1_weighted_per_method[methods.index(method)]
        f1_macro = np.random.normal(mean_f1_macro_per_method[methods.index(method)], std_f1_macro_per_method[methods.index(method)])
        if f1_macro > max_f1_macro_per_method[methods.index(method)]:
            f1_macro = max_f1_macro_per_method[methods.index(method)]
        data.append([method, seed, acc, mcc, f1_weighted, f1_macro])
df_simulated = pd.DataFrame(data, columns=['Method', 'Seed', 'Accuracy', 'MCC', 'F1_weighted', 'F1_macro'])
df_melted = pd.melt(df_simulated, id_vars=['Method', 'Seed'], var_name='Metric', value_name='Value')
df_melted.head()

fig, ax = plt.subplots(figsize=(8, 5))
colors = ['#dfa320', '#da713b', '#dd4a65', '#e46bbb', '#44b3e3', '#19b2a9', '#4b9b3f']
# colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2']
# hue: categorical variables; dodge: separate the box for different hue; width: width of the box; gap: gap between boxes
sns.boxplot(x="Metric", y="Value", hue="Method", data=df_melted, dodge='auto', width=0.9, gap=0.3, saturation=1,
            palette=colors, flierprops=dict(marker='d', markerfacecolor='black', markersize=4), linewidth=1)
plt.xticks(fontsize=10)
plt.ylim(0.5, 0.82)
plt.yticks(np.arange(0.5, 0.82, 0.02), fontsize=10)
# Add vertical lines to separate the datasets
for i in range(1, len(df_simulated["Method"].unique())):
    ax.axvline(x=i - 0.5, color='black', linestyle='-', linewidth=1)
plt.xticks(fontsize=14, fontstyle='italic')
ax.set_xlabel('')
ax.set_ylabel('', fontsize=12)
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)
ax.spines['top'].set_linewidth(1)
ax.spines['right'].set_linewidth(1)
ax.spines['bottom'].set_linewidth(1)
ax.spines['left'].set_linewidth(1)
plt.legend(loc='center right', bbox_to_anchor=(1.2, 0.5), title="", ncol=1, frameon=False, fontsize=10)
plt.legend().set_visible(False)
plt.tight_layout(rect=[0, 0, 1, 0.85]) # left, bottom, right, top
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Generate random data
np.random.seed(0)
data = [np.random.normal(loc, 0.5, 100) for loc in range(5)]
mean = [np.mean(d) for d in data]

# Define the cubehelix palette
palette = sns.cubehelix_palette(start=1, rot=0, dark=0.2, light=.75, reverse=True, n_colors=5)

# Create the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))
box = ax.boxplot(data, patch_artist=True, showfliers=False) # Draw the boxplot and manually apply the colors from the palette

# Set color for each box using the cubehelix palette， and set whiskers, caps, and medians to match each box's color
for patch, color in zip(box['boxes'], palette):
    patch.set_facecolor(color)
    patch.set_edgecolor(color)
    patch.set_linewidth(2)  # Thicker border for visibility
for whisker, color in zip(box['whiskers'], [color for color in palette for _ in (0, 1)]):  # Each box has two whiskers
    whisker.set_color(color)
    whisker.set_linewidth(2)
for cap, color in zip(box['caps'], [color for color in palette for _ in (0, 1)]):  # Each box has two caps
    cap.set_color(color)
    cap.set_linewidth(2)
for median, color in zip(box['medians'], palette):
    median.set_color(color)
    median.set_linewidth(2)
    
# add line for mean
plt.plot([1, 2, 3, 4, 5], mean, marker='o', color='gray', label='Mean', linewidth=1, markersize=4, zorder=3)

# Set the title and labels
# plt.title("Boxplot with Cubehelix Gradient Colors")
plt.xticks([1, 2, 3, 4, 5], ['10%', '30%', '50%', '70%', '90%'])
# plt.xlabel("Groups")
# plt.ylabel("Values")
plt.grid(False)
plt.show()
```
