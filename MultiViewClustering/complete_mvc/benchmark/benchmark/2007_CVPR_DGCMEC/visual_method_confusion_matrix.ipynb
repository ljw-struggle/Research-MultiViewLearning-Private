```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 生成一个新的模拟混淆矩阵，使得对角线数据更多，表示模型更准确
np.random.seed(42)
confusion_matrix_accurate = np.random.randint(0, 10, size=(10, 10))

cell_types = ["T cells", "B cells", "Monocytes", "NK cells", "Dendritic cells", 
                   "Neutrophils", "Eosinophils", "Basophils", "Megakaryocytes", "Erythrocytes"]
# 生成 10 种真实存在的细胞类型


# 增加对角线的数据，使其比非对角线的数据大
np.fill_diagonal(confusion_matrix_accurate, np.random.randint(100, 200, size=10))

# 对该新的混淆矩阵进行按行归一化
confusion_matrix_normalized_accurate = confusion_matrix_accurate / confusion_matrix_accurate.sum(axis=1, keepdims=True) * 100
print(confusion_matrix_normalized_accurate)

# 绘制新的按行归一化后的混淆矩阵热力图
plt.figure(figsize=(8, 8))

# 绘制按行归一化后的热力图，移除方块上的数字
sns.heatmap(confusion_matrix_normalized_accurate, annot=False, fmt=".2f", cmap="Blues", cbar=False, 
            vmin=0, vmax=100, cbar_kws={'format': '%.0f%%'},
            xticklabels=cell_types, yticklabels=cell_types, linewidths=1, linecolor='black')

# 把竖轴和横轴的刻度文字放到上面和右边
plt.tick_params(axis='x', top=True, labeltop=True, labelbottom=False)
plt.tick_params(axis='y', right=True, labelright=True, labelleft=False)

plt.yticks(rotation=0, fontsize=22, color = 'w')  # 红色字体
plt.xticks(rotation=90, fontsize=22)  # 横轴刻度文字旋转 90 度

# 删除横轴和竖轴的名字
plt.xlabel('')
plt.ylabel('')

# 删除标题
plt.title('')

# # 添加横轴和竖轴 10 的横线和竖线
ax = plt.gca()  # 获取当前轴
ax.axhline(y=10, color='black', linewidth=4)  # 横轴线（底部）
ax.axvline(x=10, color='black', linewidth=4)  # 最右侧竖轴线

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

# 创建一个归一化的颜色范围，从 0 到 1
norm = mpl.colors.Normalize(vmin=0, vmax=100)

# 生成一个色彩图（可以使用任何你喜欢的 colormap）
cmap = plt.get_cmap("Blues")

# 创建一个带有颜色条的figure
fig, ax = plt.subplots(figsize=(1, 6))
fig.subplots_adjust(right=0.5)

# 创建颜色条，指定归一化范围和颜色映射
cbar = fig.colorbar(mpl.cm.ScalarMappable(norm=norm, cmap=cmap), cax=ax, orientation='vertical')

# 设置颜色条的标签格式为百分比
cbar.ax.xaxis.set_ticks_position('top')
cbar.set_label('Percentage (%)', fontsize=20)

# 设置 刻度文字的大小
cbar.ax.tick_params(labelsize=20)

plt.show()
```
