```python
import os
import numpy as np
import pandas as pd

os.makedirs('./figure/comparison_with_methods/', exist_ok=True)
os.makedirs('./figure/comparison_with_methods/data_sc_multiomics/', exist_ok=True)
os.makedirs('./figure/comparison_with_methods/data_bulk_multiomics/', exist_ok=True)

def load_clustering_results_from_files(datasets, methods, method_dir_dict, data_type='sc_multiomics'):
    base_dir = f'./result/data_{data_type}'
    metric_column_map = {'acc': 'Clustering ACC', 'nmi': 'Clustering NMI', 'ari': 'Clustering ARI', 'ri': 'Clustering RI', 'asw': 'Clustering ASW', 'purity': 'Clustering Purity'}
    df_dict = {}
    for metric in metric_column_map.keys():
        df_dict[metric] = pd.DataFrame(index=datasets, columns=methods)
    for dataset in datasets:
        for method in methods:
            method_dir = method_dir_dict.get(method, method)
            result_file = os.path.join(base_dir, method_dir, dataset, 'clustering_result.csv')
            print(result_file)
            if os.path.exists(result_file):
                df_result = pd.read_csv(result_file, index_col=0)
                for metric, column_name in metric_column_map.items():
                    if column_name in df_result.columns:
                        values = df_result[column_name].values
                        values = [f'{x:.2f}' for x in values]
                        df_dict[metric].loc[dataset, method] = values
                    else:
                        print(f"Warning: Column '{column_name}' not found in {result_file}")
                        df_dict[metric].loc[dataset, method] = np.array([0, 0, 0, 0, 0])
            else:
                print(f"File not found: {result_file}")
                for metric in metric_column_map.keys():
                    df_dict[metric].loc[dataset, method] = np.array([0, 0, 0, 0, 0])
    return df_dict

def generate_dataset_summary_tables(df_results, datasets, methods, save_dir,):
    metric_names = {'acc': 'ACC', 'nmi': 'NMI', 'ari': 'ARI', 'ri': 'RI', 'asw': 'ASW', 'purity': 'Purity'}
    for dataset in datasets:
        summary_data = []
        columns = []
        for metric_key, metric_name in metric_names.items():
            columns.append(f'{metric_name}_mean')
            columns.append(f'{metric_name}_std')
        for method in methods:
            row_data = []
            for metric_key in metric_names.keys():
                values = df_results[metric_key].loc[dataset, method]
                values = [float(x) for x in values]
                mean_val = np.mean(values).round(2)
                std_val = np.std(values, ddof=0).round(2)  # ddof=1 for sample standard deviation
                row_data.extend([mean_val, std_val])
            summary_data.append(row_data)
        df_summary = pd.DataFrame(summary_data, index=methods, columns=columns)
        df_summary.to_csv(os.path.join(save_dir, f'summary_{dataset}.csv'), index=True)
        print(f"\n{dataset} dataset summary:")
        print(df_summary.round(2))
        print("\n" + "="*80 + "\n")

datasets = ['TEA', 'DOGMA', 'NEAT']
methods = ['HDUMEC', 'scMHNN', 'scMDC', 'Matilda', 'MOJITOO', 'MOFA2', 'ZINBVAE', 'VAE', 'DAE', 'AE', 'HDUR']
method_dir_dict = {'HDUMEC': 'DUCMME', 'scMHNN': 'SCMHNN', 'scMDC': 'SCMDC', 'Matilda': 'MATILDA', 
                   'MOJITOO': 'MOJITOO', 'MOFA2': 'MOFA2', 'ZINBVAE': 'ZINBVAE', 'VAE': 'VAE', 'DAE': 'DAE', 'AE': 'AE', 'HDUR': 'HDUR'}
df_results = load_clustering_results_from_files(datasets, methods, method_dir_dict, data_type='sc_multiomics')
df_results['acc'].to_csv('./figure/comparison_with_methods/data_sc_multiomics/comparison_clustering_result_acc.csv', index=True)
df_results['nmi'].to_csv('./figure/comparison_with_methods/data_sc_multiomics/comparison_clustering_result_nmi.csv', index=True)
df_results['ari'].to_csv('./figure/comparison_with_methods/data_sc_multiomics/comparison_clustering_result_ari.csv', index=True)
df_results['ri'].to_csv('./figure/comparison_with_methods/data_sc_multiomics/comparison_clustering_result_ri.csv', index=True)
df_results['asw'].to_csv('./figure/comparison_with_methods/data_sc_multiomics/comparison_clustering_result_asw.csv', index=True)
df_results['purity'].to_csv('./figure/comparison_with_methods/data_sc_multiomics/comparison_clustering_result_purity.csv', index=True)
os.makedirs('./figure/comparison_with_methods/data_sc_multiomics/dataset_summary/', exist_ok=True)
generate_dataset_summary_tables(df_results, datasets, methods, save_dir='./figure/comparison_with_methods/data_sc_multiomics/dataset_summary/')
datasets = ['BRCA', 'KIPAN', 'LGG']
methods = ['HDUMEC', 'scMHNN', 'scMDC', 'Matilda', 'MOJITOO', 'MOFA2', 'ZINBVAE', 'VAE', 'DAE', 'AE', 'HDUR']
method_dir_dict = {'HDUMEC': 'DUCMME', 'scMHNN': 'SCMHNN', 'scMDC': 'SCMDC', 'Matilda': 'MATILDA', 
                   'MOJITOO': 'MOJITOO', 'MOFA2': 'MOFA2', 'ZINBVAE': 'ZINBVAE', 'VAE': 'VAE', 'DAE': 'DAE', 'AE': 'AE', 'HDUR': 'HDUR'}
df_results = load_clustering_results_from_files(datasets, methods, method_dir_dict, data_type='bulk_multiomics')
df_results['acc'].to_csv('./figure/comparison_with_methods/data_bulk_multiomics/comparison_clustering_result_acc.csv', index=True)
df_results['nmi'].to_csv('./figure/comparison_with_methods/data_bulk_multiomics/comparison_clustering_result_nmi.csv', index=True)
df_results['ari'].to_csv('./figure/comparison_with_methods/data_bulk_multiomics/comparison_clustering_result_ari.csv', index=True)
df_results['ri'].to_csv('./figure/comparison_with_methods/data_bulk_multiomics/comparison_clustering_result_ri.csv', index=True)
df_results['asw'].to_csv('./figure/comparison_with_methods/data_bulk_multiomics/comparison_clustering_result_asw.csv', index=True)
df_results['purity'].to_csv('./figure/comparison_with_methods/data_bulk_multiomics/comparison_clustering_result_purity.csv', index=True)
os.makedirs('./figure/comparison_with_methods/data_bulk_multiomics/dataset_summary/', exist_ok=True)
generate_dataset_summary_tables(df_results, datasets, methods, save_dir='./figure/comparison_with_methods/data_bulk_multiomics/dataset_summary/')
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 14
np.random.seed(42)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf', '#1f77b4']
os.makedirs('./figure/comparison_with_methods/', exist_ok=True)

def plot_model_comparison_dataset_boxplot(metric_name, df_dataset_method_metric, save_path, prefix=''):
    datasets = df_dataset_method_metric.index.tolist()
    methods = df_dataset_method_metric.columns.tolist()
    metrics_name_dict = {
        'acc': 'Clustering ACC (%)',
        'nmi': 'Normalized Mutual Information (%)',
        'ari': 'Adjusted Rand Index (%)',
        'ri': 'Rand Index (%)',
        'asw': 'Average Silhouette Width (%)',
        'purity': 'Purity Score (%)',
    }
    fig = plt.figure(figsize=(10, 6))
    labels = df_dataset_method_metric.columns.tolist()
    def set_boxplot_colors(bp, color_list):
        for patch, color in zip(bp['boxes'], color_list):
            patch.set_facecolor(color)
        plt.setp(bp['medians'], color='black')
        plt.setp(bp['whiskers'], color='black')
    
    dataset_1_values = [[eval(value) for value in eval(df_dataset_method_metric.loc[datasets[0], method])] for method in methods]
    positions_1 = list(range(1, len(methods) + 1))
    boxplot_1 = plt.boxplot(dataset_1_values, patch_artist=True, labels=labels, positions=positions_1, widths=0.6, showfliers=False)
    set_boxplot_colors(boxplot_1, colors[:len(methods)])
    
    dataset_2_values = [[eval(value) for value in eval(df_dataset_method_metric.loc[datasets[1], method])] for method in methods]
    positions_2 = list(range(len(methods) + 2, len(methods) * 2 + 2))
    boxplot_2 = plt.boxplot(dataset_2_values, patch_artist=True, labels=labels, positions=positions_2, widths=0.6, showfliers=False)
    set_boxplot_colors(boxplot_2, colors[:len(methods)])
    
    dataset_3_values = [[eval(value) for value in eval(df_dataset_method_metric.loc[datasets[2], method])] for method in methods]
    positions_3 = list(range(len(methods) * 2 + 3, len(methods) * 3 + 3))
    boxplot_3 = plt.boxplot(dataset_3_values, patch_artist=True, labels=labels, positions=positions_3, widths=0.6, showfliers=False)
    set_boxplot_colors(boxplot_3, colors[:len(methods)])
    x_position = [(positions_1[0] + positions_1[-1]) / 2, (positions_2[0] + positions_2[-1]) / 2, (positions_3[0] + positions_3[-1]) / 2]
    
    plt.xticks(x_position, datasets, fontsize=14)
    plt.xlim(0, positions_3[-1] + 1)
    plt.yticks(fontsize=14)
    plt.ylim(0, 100)
    # plt.xlabel('Datasets', fontsize=14)
    plt.ylabel(metrics_name_dict[metric_name], fontsize=14)
    # plt.legend(boxplot_1["boxes"], labels, loc='lower right', bbox_to_anchor=(1.10, 0.0), fontsize=14, borderaxespad=0)
    # plt.legend(boxplot_1["boxes"], labels, loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=len(labels), fontsize=14, borderaxespad=0, fancybox=True, shadow=True)
    # plt.legend(boxplot_1["boxes"], labels, loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=len(labels), fontsize=14, borderaxespad=0, frameon=False)
    # plt.legend(boxplot_1["boxes"], labels, loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=len(labels), fontsize=14, borderaxespad=0, frameon=False)
    # plt.legend(boxplot_1["boxes"], [label.upper() for label in labels], loc='upper center', bbox_to_anchor=(0.5, -0.12), 
    #            ncol=4, fontsize=11, borderaxespad=0.5, frameon=False,
    #            columnspacing=1.0, handlelength=3.5, handleheight=1.2, handletextpad=0.5)
    plt.plot([(positions_1[-1] + positions_2[0]) / 2, (positions_1[-1] + positions_2[0]) / 2], plt.ylim(), 'k--', lw=1)
    plt.plot([(positions_2[-1] + positions_3[0]) / 2, (positions_2[-1] + positions_3[0]) / 2], plt.ylim(), 'k--', lw=1)
    plt.tight_layout()
    plt.savefig(os.path.join(save_path, f'{prefix}_boxplot_{metric_name}.png'), dpi=400, bbox_inches='tight')
    # plt.show()
    plt.close()

def plot_model_comparison_dataset_barplot(metric_name, df_dataset_method_metric, save_path, prefix=''):
    datasets = df_dataset_method_metric.index.tolist()
    methods = df_dataset_method_metric.columns.tolist()
    metrics_name_dict = {
        'acc': 'Clustering ACC (%)',
        'nmi': 'Normalized Mutual Information (%)',
        'ari': 'Adjusted Rand Index (%)',
        'ri': 'Rand Index (%)',
        'asw': 'Average Silhouette Width (%)',
        'purity': 'Purity Score (%)',
    }
    fig = plt.figure(figsize=(10, 6))
    labels = df_dataset_method_metric.columns.tolist()
    dataset_1_means = []
    dataset_1_stds = []
    for method in methods:
        values = df_dataset_method_metric.loc[datasets[0], method]
        values = [eval(x) for x in eval(values)]
        dataset_1_means.append(np.mean(values))
        dataset_1_stds.append(np.std(values))
    positions_1 = list(range(1, len(methods) + 1))
    bars_1 = plt.bar(positions_1, dataset_1_means, width=0.6, yerr=dataset_1_stds, 
                     capsize=3, color=colors[:len(methods)], label=datasets[0], alpha=0.8, edgecolor='black', linewidth=1)
    dataset_2_means = []
    dataset_2_stds = []
    for method in methods:
        values = df_dataset_method_metric.loc[datasets[1], method]
        values = [eval(x) for x in eval(values)]
        dataset_2_means.append(np.mean(values))
        dataset_2_stds.append(np.std(values))
    positions_2 = list(range(len(methods) + 2, len(methods) * 2 + 2))
    bars_2 = plt.bar(positions_2, dataset_2_means, width=0.6, yerr=dataset_2_stds, 
                     capsize=3, color=colors[:len(methods)], label=datasets[1], alpha=0.8, edgecolor='black', linewidth=1)
    dataset_3_means = []
    dataset_3_stds = []
    for method in methods:
        values = df_dataset_method_metric.loc[datasets[2], method]
        values = [eval(x) for x in eval(values)]
        dataset_3_means.append(np.mean(values))
        dataset_3_stds.append(np.std(values))
    positions_3 = list(range(len(methods) * 2 + 3, len(methods) * 3 + 3))
    bars_3 = plt.bar(positions_3, dataset_3_means, width=0.6, yerr=dataset_3_stds, 
                     capsize=3, color=colors[:len(methods)], label=datasets[2], alpha=0.8, edgecolor='black', linewidth=1)
    x_position = [(positions_1[0] + positions_1[-1]) / 2, (positions_2[0] + positions_2[-1]) / 2, (positions_3[0] + positions_3[-1]) / 2]
    plt.xticks(x_position, datasets, fontsize=14)
    plt.xlim(0, positions_3[-1] + 1)
    plt.yticks(fontsize=14)
    plt.ylim(0, 100)
    plt.ylabel(metrics_name_dict[metric_name], fontsize=14)
    # plt.legend(bars_1, [label.upper() for label in labels], loc='upper center', bbox_to_anchor=(0.5, -0.12), 
    #            ncol=4, fontsize=11, borderaxespad=0.5, frameon=False,
    #            columnspacing=1.0, handlelength=1.5, handletextpad=0.5)
    ymin, ymax = plt.ylim()
    plt.plot([(positions_1[-1] + positions_2[0]) / 2, (positions_1[-1] + positions_2[0]) / 2], [ymin, ymax], 'k--', lw=1)
    plt.plot([(positions_2[-1] + positions_3[0]) / 2, (positions_2[-1] + positions_3[0]) / 2], [ymin, ymax], 'k--', lw=1)
    plt.tight_layout()
    plt.savefig(os.path.join(save_path, f'{prefix}_barplot_{metric_name}.png'), dpi=400, bbox_inches='tight')
    # plt.show()
    plt.close()
    
def plot_legend_only(methods, save_path):
    fig = plt.figure(figsize=(12, 1))
    ax = fig.add_subplot(111)
    handles = []
    for i, (method, color) in enumerate(zip(methods, colors[:len(methods)])):
        patch = plt.Rectangle((0, 0), 1, 1, facecolor=color, edgecolor='black', linewidth=1, alpha=0.8)
        handles.append(patch)
    legend = ax.legend(handles, methods, loc='center', ncol=len(methods), 
                       fontsize=14, frameon=False, fancybox=True, shadow=False,
                       columnspacing=1.5, handlelength=2.0, handletextpad=0.8)
    ax.axis('off')
    plt.tight_layout()
    plt.savefig(os.path.join(save_path, f'legend.png'), dpi=400, bbox_inches='tight', pad_inches=0.1)
    # plt.show()
    plt.close()
    
df_metrics_acc = pd.read_csv('./figure/comparison_with_methods/data_sc_multiomics/comparison_clustering_result_acc.csv', index_col=0)
df_metrics_nmi = pd.read_csv('./figure/comparison_with_methods/data_sc_multiomics/comparison_clustering_result_nmi.csv', index_col=0)
df_metrics_ari = pd.read_csv('./figure/comparison_with_methods/data_sc_multiomics/comparison_clustering_result_ari.csv', index_col=0)
df_metrics_ri = pd.read_csv('./figure/comparison_with_methods/data_sc_multiomics/comparison_clustering_result_ri.csv', index_col=0)
df_metrics_asw = pd.read_csv('./figure/comparison_with_methods/data_sc_multiomics/comparison_clustering_result_asw.csv', index_col=0)
df_metrics_purity = pd.read_csv('./figure/comparison_with_methods/data_sc_multiomics/comparison_clustering_result_purity.csv', index_col=0)
plot_model_comparison_dataset_boxplot('acc', df_dataset_method_metric=df_metrics_acc, save_path='./figure/comparison_with_methods/', prefix='sc_multi_omics')
plot_model_comparison_dataset_boxplot('nmi', df_dataset_method_metric=df_metrics_nmi, save_path='./figure/comparison_with_methods/', prefix='sc_multi_omics')
plot_model_comparison_dataset_boxplot('ari', df_dataset_method_metric=df_metrics_ari, save_path='./figure/comparison_with_methods/', prefix='sc_multi_omics')
plot_model_comparison_dataset_boxplot('ri', df_dataset_method_metric=df_metrics_ri, save_path='./figure/comparison_with_methods/', prefix='sc_multi_omics')
plot_model_comparison_dataset_boxplot('asw', df_dataset_method_metric=df_metrics_asw, save_path='./figure/comparison_with_methods/', prefix='sc_multi_omics')
plot_model_comparison_dataset_boxplot('purity', df_dataset_method_metric=df_metrics_purity, save_path='./figure/comparison_with_methods/', prefix='sc_multi_omics')
plot_model_comparison_dataset_barplot('acc', df_dataset_method_metric=df_metrics_acc, save_path='./figure/comparison_with_methods/', prefix='sc_multi_omics')
plot_model_comparison_dataset_barplot('nmi', df_dataset_method_metric=df_metrics_nmi, save_path='./figure/comparison_with_methods/', prefix='sc_multi_omics')
plot_model_comparison_dataset_barplot('ari', df_dataset_method_metric=df_metrics_ari, save_path='./figure/comparison_with_methods/', prefix='sc_multi_omics')
plot_model_comparison_dataset_barplot('ri', df_dataset_method_metric=df_metrics_ri, save_path='./figure/comparison_with_methods/', prefix='sc_multi_omics')
plot_model_comparison_dataset_barplot('asw', df_dataset_method_metric=df_metrics_asw, save_path='./figure/comparison_with_methods/', prefix='sc_multi_omics')
plot_model_comparison_dataset_barplot('purity', df_dataset_method_metric=df_metrics_purity, save_path='./figure/comparison_with_methods/', prefix='sc_multi_omics')

df_metrics_acc = pd.read_csv('./figure/comparison_with_methods/data_bulk_multiomics/comparison_clustering_result_acc.csv', index_col=0) 
df_metrics_nmi = pd.read_csv('./figure/comparison_with_methods/data_bulk_multiomics/comparison_clustering_result_nmi.csv', index_col=0)
df_metrics_ari = pd.read_csv('./figure/comparison_with_methods/data_bulk_multiomics/comparison_clustering_result_ari.csv', index_col=0)
df_metrics_ri = pd.read_csv('./figure/comparison_with_methods/data_bulk_multiomics/comparison_clustering_result_ri.csv', index_col=0)
df_metrics_asw = pd.read_csv('./figure/comparison_with_methods/data_bulk_multiomics/comparison_clustering_result_asw.csv', index_col=0)
df_metrics_purity = pd.read_csv('./figure/comparison_with_methods/data_bulk_multiomics/comparison_clustering_result_purity.csv', index_col=0)
plot_model_comparison_dataset_boxplot('acc', df_dataset_method_metric=df_metrics_acc, save_path='./figure/comparison_with_methods/', prefix='bulk_multi_omics')
plot_model_comparison_dataset_boxplot('nmi', df_dataset_method_metric=df_metrics_nmi, save_path='./figure/comparison_with_methods/', prefix='bulk_multi_omics')
plot_model_comparison_dataset_boxplot('ari', df_dataset_method_metric=df_metrics_ari, save_path='./figure/comparison_with_methods/', prefix='bulk_multi_omics')
plot_model_comparison_dataset_boxplot('ri', df_dataset_method_metric=df_metrics_ri, save_path='./figure/comparison_with_methods/', prefix='bulk_multi_omics')
plot_model_comparison_dataset_boxplot('asw', df_dataset_method_metric=df_metrics_asw, save_path='./figure/comparison_with_methods/', prefix='bulk_multi_omics')
plot_model_comparison_dataset_boxplot('purity', df_dataset_method_metric=df_metrics_purity, save_path='./figure/comparison_with_methods/', prefix='bulk_multi_omics')
plot_model_comparison_dataset_barplot('acc', df_dataset_method_metric=df_metrics_acc, save_path='./figure/comparison_with_methods/', prefix='bulk_multi_omics')
plot_model_comparison_dataset_barplot('nmi', df_dataset_method_metric=df_metrics_nmi, save_path='./figure/comparison_with_methods/', prefix='bulk_multi_omics')
plot_model_comparison_dataset_barplot('ari', df_dataset_method_metric=df_metrics_ari, save_path='./figure/comparison_with_methods/', prefix='bulk_multi_omics')
plot_model_comparison_dataset_barplot('ri', df_dataset_method_metric=df_metrics_ri, save_path='./figure/comparison_with_methods/', prefix='bulk_multi_omics')
plot_model_comparison_dataset_barplot('asw', df_dataset_method_metric=df_metrics_asw, save_path='./figure/comparison_with_methods/', prefix='bulk_multi_omics')
plot_model_comparison_dataset_barplot('purity', df_dataset_method_metric=df_metrics_purity, save_path='./figure/comparison_with_methods/', prefix='bulk_multi_omics')

plot_legend_only(methods, save_path='./figure/comparison_with_methods/')
```
