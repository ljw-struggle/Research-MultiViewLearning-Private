```python
import requests

def query_biomart(xml_query):
    response = requests.post("http://www.ensembl.org/biomart/martservice?query=", data=xml_query)
    return response.text if response.ok else "Error: Unable to fetch data"

xml_query = """
<?xml version="1.0" encoding="UTF-8"?>
<!DOCTYPE Query>
<Query virtualSchemaName = "default" formatter = "TSV" header = "0" uniqueRows = "0" count = "" datasetConfigVersion = "0.6" >
    <Dataset name = "hsapiens_gene_ensembl" interface = "default" >
        <Attribute name = "ensembl_gene_id" />
        <Attribute name = "external_gene_name" />
    </Dataset>
</Query>
"""
biomart_data = query_biomart(xml_query)
print(biomart_data)
```
