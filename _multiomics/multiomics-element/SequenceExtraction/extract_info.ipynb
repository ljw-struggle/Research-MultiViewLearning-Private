```python
import pandas as pd
from Bio import Entrez, SeqIO

Entrez.email = "ljwstruggle@gmail.com"

def fetch_gene_info(gene_id):
    with Entrez.efetch(db="gene", id=gene_id, retmode="xml") as handle:
        gene_record = Entrez.read(handle)
    return gene_record

def fetch_protein_info(protein_id):
    with Entrez.efetch(db="protein", id=protein_id, rettype="fasta", retmode="text") as handle:
        protein_record = SeqIO.read(handle, "fasta")
    return protein_record
```
