```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Calculate the distribution of the sequence length in the dataset

```python
homo_tf = pd.read_csv('./TFLinkDataset/ProcessedData/HomoSapiens/protein.seq', header=None, sep='\t')
mus_tf = pd.read_csv('./TFLinkDataset/ProcessedData/MusMusculus/protein.seq', header=None, sep='\t')
homo_tf_length = [len(sequence) for sequence in homo_tf[1].values]
mus_tf_length = [len(sequence) for sequence in mus_tf[1].values]
```

```python
# Plot the distribution of TF length of HomoSapiens
plt.style.use('seaborn')
plt.figure(figsize=(10, 5), dpi=100)
group_num = int((max(homo_tf_length) - min(homo_tf_length)) / 20)
plt.hist(homo_tf_length, bins=group_num)
plt.xlim(0, 3000)
plt.ylim(0, 80)
plt.xticks(range(0, 3001)[::1000] ,fontsize=5)
plt.yticks(fontsize=5)
plt.xlabel("length")
plt.ylabel("num")
plt.grid(linestyle="--", alpha=0.5)
plt.show()
```

```python
# Plot the distribution of TF length of MusMusculus
plt.style.use('seaborn')
plt.figure(figsize=(10, 5), dpi=100)
group_num = int((max(mus_tf_length) - min(mus_tf_length)) / 20)
plt.hist(mus_tf_length, bins=group_num)
plt.xlim(0, 3000)
plt.ylim(0, 80)
plt.xticks(range(0, 3001)[::1000] ,fontsize=5)
plt.yticks(fontsize=5)
plt.xlabel("length")
plt.ylabel("num")
plt.grid(linestyle="--", alpha=0.5)
plt.show()
```
